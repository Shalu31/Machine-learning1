```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['difficulty'] = pd.to_numeric(df['difficulty'], errors='coerce')
df['sentiment_score'] = pd.to_numeric(df['sentiment_score'], errors='coerce')
```

```python
df.info()
```

```python
filtered_df = df[
    (df['attendance_required'] == "No") &
    (df['difficulty'] <= 2) &
    (df['rating'] >= 4.0)
]
```

```python
top_courses = filtered_df.sort_values(
    by=['sentiment_score', 'rating'],
    ascending=False
).head(5)
```

```python
print(top_courses[['course_code', 'course_name', 'rating', 'difficulty', 'sentiment_score', 'is_online']])
```
